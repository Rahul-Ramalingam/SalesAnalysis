```python
ndf = pd.read_csv('one_year_sales.csv')
#print(len(ndf['Order ID']))
ndf.head()
```

### Best month for sales.

```python
na = ndf[ndf.isna().any(axis=1)]
print(len(na['Order ID']))
```

```python
ndf = ndf.dropna(how='all')
```

```python
temp_df = ndf[ndf['Order Date'].str[:2] == 'Or']
temp_df.head()
ndf = ndf[ndf['Order Date'].str[:2] != 'Or']
```

```python
#add month column
# date = ndf['Order Date'][4]
# date[:2]
ndf['month'] = ndf['Order Date'].str[:2]
ndf['month'] = ndf['month'].astype('int32')
```

```python
#add sales column
ndf['Quantity Ordered'] = pd.to_numeric(ndf['Quantity Ordered'])
ndf['Price Each'] = pd.to_numeric(ndf['Price Each'])
ndf['sales'] = ndf['Quantity Ordered']*ndf['Price Each']
```

```python
ndf.head()
```

```python
sales_df = ndf.groupby('month').sum()
maxSales = max(sales_df['sales'])
max_sales = sales_df.loc[sales_df['sales'] == maxSales]
max_sales
#print(f"The maximum sales is on the month {max_sales['month'][0]}")
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
sales = sales_df['sales']
plt.bar(months,sales)
plt.xticks(months)
plt.xlabel("Months")
plt.ylabel("Sales in Million")
plt.show()
```

### City that has maximum sales 

```python
ndf
```

```python
#make a city column
ndf['Purchase Address'][1].split(',')[1]
ndf['Purchase Address'][1].split(', ')[2].split(' ')[0]
ndf['city'] = ndf['Purchase Address'].apply(lambda x : f"{x.split(',')[1]},{x.split(', ')[2].split(' ')[0]}")
ndf
```

```python
maxSales_city = ndf.groupby('city').sum()
maxSales_city['sales'] = maxSales_city['sales'].astype('int')
maxSalescity1 = max(maxSales_city['sales'])
maxSalescity = maxSales_city.loc[maxSales_city['sales']==maxSalescity1]
maxSalescity
```

```python
#to keep the city in correct order
city = [city for city,df in ndf.groupby('city')]
sales = maxSales_city['sales']
plt.bar(city,sales)
plt.xticks(city,rotation='vertical')
plt.xlabel("Cities")
plt.ylabel("Sales in Million")
plt.show()
```

### The time the products to be advertised

```python
#convert the order date column to datetime object
ndf['Order Date'] = pd.to_datetime(ndf['Order Date'])
#seperate hours and minutes
ndf['hour'] = ndf['Order Date'].dt.hour
ndf
```

```python
#visuualize
#x-axis hours
hours = [hour for hour,df in ndf.groupby('hour')]
#y-axis count of items sold in each hour
count = ndf.groupby(['hour']).count()
plt.plot(hours,count)
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Items Sold')
plt.grid()
plt.show()
```

##### From the graph the maximun items sold in 11:00 a.m - 12:00 p.m and 7:00 p.m

## The Products Sold together

```python
#find orders which has multiple products
mul_orders = ndf[ndf['Order ID'].duplicated(keep=False)]
mul_orders
```

```python
#create  a column by combining the products
mul_orders['grouped'] = mul_orders.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
mul_orders = mul_orders[['Order ID','grouped']].drop_duplicates()
mul_orders.head()
```

```python
#count the combinations
from itertools import combinations
from collections import Counter
count = Counter()

for i in mul_orders['grouped']:
    li = i.split(',')
    count.update(Counter(combinations(li,2)))
    
count.most_common(10)
```

##### Iphone and lightning charging cable are the most sold combination

### The product sold the most

```python
products = ndf.groupby('Product')
products.sum()
```

```python
prods = [prod for prod,df in ndf.groupby('Product')]
quantity = products.sum()['Quantity Ordered']

plt.bar(prods,quantity)
plt.xticks(prods,rotation='vertical',size=9)
plt.xlabel('Products')
plt.ylabel('Quantity')
plt.show()
```

##### AAA batteries sold the most

```python
prices = ndf.groupby('Product').mean()['Price Each']
prices
```

```python
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(prods,quantity)
ax2.plot(prods,prices,color='r')

ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity',color='blue')
ax2.set_ylabel('Price',color='r')
ax1.set_xticklabels(prods,rotation='vertical',size=9)
plt.show()
```

#### The lower the price the higher the product sold
